```python
def visible_trees(forest, i, j):
    # check if the current tree is on the edge of the grid
    if i == 0 or i == len(forest) - 1 or j == 0 or j == len(forest[0]) - 1:
        return 1

    # check if the current tree is taller than its neighbors
    if forest[i][j] >= forest[i - 1][j] and forest[i][j] >= forest[i + 1][j] and forest[i][j] >= forest[i][j - 1] and forest[i][j] >= forest[i][j + 1]:
        return 1

    # the current tree is not visible, so we check its neighbors recursively
    return visible_trees(forest, i - 1, j) + visible_trees(forest, i + 1, j) + visible_trees(forest, i, j - 1) + visible_trees(forest, i, j + 1)

```

```python
def visible_trees(forest):
    # initialize a set to keep track of which trees have been checked
    checked = set()

    num_visible = 0

    for i in range(len(forest)):
        for j in range(len(forest[0])):
            # check if the current tree is on the edge of the grid
            if i == 0 or i == len(forest) - 1 or j == 0 or j == len(forest[0]) - 1:
                num_visible += 1
                checked.add((i, j))
                continue

            # check if the current tree is taller than its neighbors
            if forest[i][j] >= forest[i - 1][j] and forest[i][j] >= forest[i + 1][j] and forest[i][j] >= forest[i][j - 1] and forest[i][j] >= forest[i][j + 1]:
                num_visible += 1
                checked.add((i, j))
                continue

            # check if the current tree has already been checked
            if (i, j) in checked:
                continue

            # the current tree is not visible, so we check its neighbors
            for di in [-1, 0, 1]:
                for dj in [-1, 0, 1]:
                    if di == 0 and dj == 0:
                        continue
                    if (i + di, j + dj) in checked:
                        continue
                    if forest[i + di][j + dj] > forest[i][j]:
                        checked.add((i, j))
                        break

    return num_visible

# read the input from the file
with open("test.txt", "r") as f:
    lines = f.readlines()

# convert the input into a 2D array
forest = [list(map(int, line.strip())) for line in lines]

# count the number of visible trees
num_visible = visible_trees(forest)

print(num_visible)
```

```python
# define the input grid as a 2D list
grid = [    [30373, 25512, 65332, 33549, 35390],
    [47359, 35499, 52324, 29485, 47382],
    [34542, 52832, 49353, 23475, 39495],
    [39443, 29485, 34598, 23498, 29384],
    [48392, 29384, 29485, 23479, 49328]
]

# create a new list to store the results
result = [[None] * len(grid[0]) for _ in range(len(grid))]

# define a function to get the trees in a given direction from the current tree to the edge of the forest
def get_trees_in_direction(grid, i, j, di, dj):
    trees = []

    while i >= 0 and i < len(grid) and j >= 0 and j < len(grid[0]):
        i += di
        j += dj
        trees.append(grid[i][j])

    return trees

# loop over each tree in the forest
for i in range(len(grid)):
    for j in range(len(grid[0])):
        # get the trees in each direction from the current tree to the edge of the forest
        up = get_trees_in_direction(grid, i, j, -1, 0)
        down = get_trees_in_direction(grid, i, j, 1, 0)
        left = get_trees_in_direction(grid, i, j, 0, -1)
        right = get_trees_in_direction(grid, i, j, 0, 1)

        # check if the current tree is on the edge of the grid
        if i == 0 or i == len(grid) - 1 or j == 0 or j == len(grid[0]) - 1:
            result[i][j] = True
            continue

        # check if the current tree is greater than all the trees in each direction
        if grid[i][j] > max(grid[up[0][0]][up[0][1]], grid[down[0][0]][down[0][1]], grid[left[0][0]][left[0][1]], grid[right[0][0]][right[0][1]]):
            result[i][j] = True

print(f"{len(grid)} x {len(grid[0])} grid")

get_trees_in_direction(grid, 3, 2, 1, 0)
```
