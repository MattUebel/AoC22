```python
bps = {}
data = open("test.txt").read().splitlines()
for bp in data:
    bp = bp.split(": ")
    bp_num = bp[0].split()[1]
    bps[bp_num] = {}
    for robot in bp[1].split("."):
        if len(robot) == 0:
            continue
        robot_type = robot.split()[1]
        bps[bp_num][robot_type] = {"costs": {}}
        if robot_type in ["ore", "clay"]:
            bps[bp_num][robot_type]["costs"]["ore"] = int(robot.split()[-2])
        elif robot_type == "obsidian":
            bps[bp_num][robot_type]["costs"]["clay"] = int(robot.split()[-2])
            bps[bp_num][robot_type]["costs"]["ore"] = int(robot.split()[-5])
        elif robot_type == "geode":
            bps[bp_num][robot_type]["costs"]["ore"] = int(robot.split()[-5])
            bps[bp_num][robot_type]["costs"]["obsidian"] = int(robot.split()[-2])
from pprint import pprint
pprint(bps)
```

```python
def find_best_bp(bps):
    best_bp = None
    max_geode_count = 0

    # Iterate through each BP in the bps dictionary
    for bp_num, robots in bps.items():
        # Initialize counters for each resource and robot type
        ore_count = 0
        clay_count = 0
        obsidian_count = 0
        geode_count = 0
        ore_robot_count = 1
        clay_robot_count = 0
        obsidian_robot_count = 0
        geode_robot_count = 0
        ore_robot_max = 2
        clay_robot_max = 3

        minute = 1
        while minute <=24:
            print("starting minute", minute)
            build_geode_robot = False
            build_obsidian_robot = False
            build_clay_robot = False
            build_ore_robot = False

            # build geode robot if possible
            if obsidian_count >= robots["geode"]["costs"]["obsidian"] and ore_count >= robots["geode"]["costs"]["ore"]:
                print("building geode robot")
                build_geode_robot = True
                obsidian_count -= robots["geode"]["costs"]["obsidian"]
                ore_count -= robots["geode"]["costs"]["ore"]
            
            # build obsidian robot if possible
            elif clay_count >= robots["obsidian"]["costs"]["clay"] and ore_count >= robots["obsidian"]["costs"]["ore"]:
                print("building obsidian robot")
                build_obsidian_robot = True
                clay_count -= robots["obsidian"]["costs"]["clay"]
                ore_count -= robots["obsidian"]["costs"]["ore"]

            # build clay robot if possible
            elif ore_count >= robots["clay"]["costs"]["ore"] and clay_robot_count < clay_robot_max:
                print("building clay robot")
                build_clay_robot = True
                ore_count -= robots["clay"]["costs"]["ore"]
            
            # build ore robot if possible
            if ore_count >= robots["ore"]["costs"]["ore"] and ore_robot_count < ore_robot_max:
                print("building ore robot")
                build_ore_robot = True
                ore_count -= robots["ore"]["costs"]["ore"]

            # increment ore count
            print(f"adding {ore_robot_count} ore to {ore_count}")
            ore_count += ore_robot_count
            print(f"adding {clay_robot_count} clay to {clay_count}")
            clay_count += clay_robot_count
            print(f"adding {obsidian_robot_count} obsidian to {obsidian_count}")
            obsidian_count += obsidian_robot_count
            print(f"adding {geode_robot_count} geode to {geode_count}")
            geode_count += geode_robot_count

            if build_geode_robot:
                geode_robot_count += 1
            if build_obsidian_robot:
                obsidian_robot_count += 1
            if build_clay_robot:
                clay_robot_count += 1
            
            minute += 1
        
        print(f"BP {bp_num} can make {geode_count} geodes in 24 minutes")


find_best_bp(bps)

```
