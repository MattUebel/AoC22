```python
# Store the heightmap as a list of strings
heightmap = ["Sabqponm", "abcryxxl", "accszExk", "acctuvwj", "abdefghi"]

# the heights are represented by the letters in the heightmap
# a is the lowest and z is the highest

# Get the dimensions of the grid
rows = len(heightmap)
cols = len(heightmap[0])

# Define a function to check if a given position is within the bounds of the grid
def is_valid(row, col):
    return 0 <= row < rows and 0 <= col < cols

# Define a function to calculate the distance between two positions
def manhattan_distance(row1, col1, row2, col2):
    return abs(row1 - row2) + abs(col1 - col2)

# Define a function to check if a given position is the destination
def is_destination(row, col):
    return heightmap[row][col] == "E"

# function that determines if a position movement is valid
# a move is valid if it is within the bounds of the grid
# and if the height of the new position is at most 1 unit higher than the current position
def is_valid_move(row, col, new_row, new_col):
    if is_valid(new_row, new_col):
        print(f"{row}, {col} -> {new_row}, {new_col}")
        return ord(heightmap[new_row][new_col]) - ord(heightmap[row][col]) <= 1
        #return abs(ord(heightmap[row][col]) - ord(heightmap[new_row][new_col])) <= 1
    print(f"{row}, {col} -> {new_row}, {new_col} is not valid")
    return False

# recursive function that finds the shortest path from the current position to the destination
def find_shortest_path(row, col, visited):
    # if the current position is the destination, return the empty string
    if is_destination(row, col):
        print("found destination")
        return ""
    # if the current position has already been visited, return None
    if visited[row][col]:
        return None
    # mark the current position as visited
    visited[row][col] = True
    # define a variable to store the shortest path
    shortest_path = None
    # loop through all possible moves
    for new_row, new_col in [(row - 1, col), (row + 1, col), (row, col - 1), (row, col + 1)]:
        print("new_row: ", new_row, "new_col: ", new_col)
        # if the move is valid, find the shortest path from the new position to the destination
        if is_valid_move(row, col, new_row, new_col):
            print("entering recursion")
            path = find_shortest_path(new_row, new_col, visited)
            # if a path was found, check if it is the shortest path found so far
            if path is not None:
                if shortest_path is None or len(path) < len(shortest_path):
                    shortest_path = path
    # if a shortest path was found, add the current position to the path
    if shortest_path is not None:
        shortest_path = heightmap[row][col] + shortest_path
    # return the shortest path found so far
    return shortest_path

# find the starting position
for row in range(rows):
    for col in range(cols):
        if heightmap[row][col] == "S":
            start_row = row
            start_col = col
            break

print(start_row, start_col)

# find the shortest path
visited = [[False] * cols for _ in range(rows)]
print(visited)
path = find_shortest_path(start_row, start_col, visited)

# print the path
print(path)


```

```python
import heapq

class HeightMap:
    def __init__(self, map_str):
        self.map = [list(line) for line in map_str.split('\n') if line]
        self.start = None
        self.end = None
        self.width = len(self.map[0])
        self.height = len(self.map)
        for y, row in enumerate(self.map):
            for x, cell in enumerate(row):
                if cell == 'S':
                    self.start = (x, y)
                elif cell == 'E':
                    self.end = (x, y)
    
    def get_elevation(self, x, y):
        # check if at start
        if (x, y) == self.start:
            return 0
        return ord(self.map[y][x]) - ord('a')
    
    def is_valid_move(self, x, y, dx, dy):
        x2 = x + dx
        y2 = y + dy
        if not (0 <= x2 < self.width and 0 <= y2 < self.height):
            return False
        elevation = self.get_elevation(x, y)
        elevation2 = self.get_elevation(x2, y2)
        print(f"elvation is {elevation}, elevation2 is {elevation2}")
        return abs(elevation - elevation2) <= 1
    
    def get_moves(self, x, y):
        moves = []
        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                if abs(dx) + abs(dy) == 1 and self.is_valid_move(x, y, dx, dy):
                    moves.append((dx, dy))
        return moves


class Climber:
    def __init__(self, heightmap):
        self.heightmap = heightmap
        self.start = heightmap.start
        self.end = heightmap.end
        self.visited = set()
    
    def search(self):
        queue = [(0, self.start)]
        while queue:
            path, (x, y) = heapq.heappop(queue)
            print(f"path is {path}, x is {x}, y is {y}")
            if (x, y) == self.end:
                return path
            if (x, y) in self.visited:
                continue
            self.visited.add((x, y))
            for dx, dy in self.heightmap.get_moves(x, y):
                print(f"making move {dx}, {dy}")
                new_path = path + 1
                heapq.heappush(queue, (new_path, (x + dx, y + dy)))
        print("done searching, no path")
        return None


heightmap = HeightMap(open('input.txt').read())
print(heightmap.start)
print(heightmap.end)

climber = Climber(heightmap)
print(climber.search())
print(climber.visited)
```
